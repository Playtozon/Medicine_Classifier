```python
use_cols = [col for col in df.columns if col.startswith('use')]
side_cols = [col for col in df.columns if col.startswith('sideEffect')]

df['uses'] = df[use_cols].apply(
    lambda row: ' '.join([
        str(x).strip() for x in row 
        if pd.notna(x) and str(x).strip().lower() != 'none' and str(x).strip() != ''
    ]),
    axis=1
)

df['side_effects'] = df[side_cols].apply(
    lambda row: ' '.join([
        str(x).strip() for x in row 
        if pd.notna(x) and str(x).strip().lower() != 'none' and str(x).strip() != ''
    ]),
    axis=1
)

def build_description(row):
    parts = []
    
    if row['uses']:
        parts.append("Uses: " + row['uses'])
        
    if row['side_effects']:
        parts.append("Side effects: " + row['side_effects'])
        
    return ". ".join(parts)

df['Description'] = df.apply(build_description, axis=1)
```

```python
sub_cols = [col for col in df.columns if col.startswith('substitute')]

df['substrates'] = df[sub_cols].apply(
    lambda row: ' '.join([
        str(x).strip() for x in row 
        if pd.notna(x) and str(x).strip().lower() != 'none' and str(x).strip() != ''
    ]),
    axis=1
)
```
